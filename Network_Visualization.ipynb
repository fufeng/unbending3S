```python
import networkx as nx
import matplotlib.pyplot as plt
```

# paint some network structures

### cycle ( $N = 30$ )

```python
# initialize a cycle with 30 nodes
cycle = nx.cycle_graph(30)

# the size of figure
plt.figure(figsize=(4, 4))

# draw the figure
nx.draw(cycle, node_color='red', node_size=25, linewidths=1, font_weight='bold', pos=nx.circular_layout(cycle))

# save the figure
#plt.savefig('cycle.svg', format='svg')

plt.show()
```

### lattice ( $N=100$ )

```python
# initialize a lattice with 100 nodes
lattice = nx.grid_2d_graph(10, 10)

# the size of figure
plt.figure(figsize=(4, 4))

# adjust the position
pos = {(i, j): (i, -j) for i, j in lattice.nodes()}

# draw the figure
nx.draw(lattice, pos, node_color='red', node_size=25,width=1)

# save the figure
#plt.savefig('lattice.svg', format='svg')

plt.show()
```

### random regular graph ( $N = 50$, $k = 3$ )

```python
# initialize a random regular graph with 50 nodes, and the average degree is 3
random_regular = nx.random_regular_graph(3, 50)

# the size of figure
plt.figure(figsize=(4, 4))

# adjust the position
pos = nx.spring_layout(random_regular, seed=42)

# draw the figure
nx.draw(random_regular, pos, node_color='red', node_size=25, linewidths=1, font_weight='bold')

# save the figure
#plt.savefig('random regular graph.svg', format='svg')

plt.show()
```

### random graph ( $N = 40$, $k = 4$ )

```python
# initialize a random graph with 40 nodes, and the average degree is 4
n = 40  # the number of nodes
G = nx.erdos_renyi_graph(n, 4/(n-1))

# the size of figure
plt.figure(figsize=(4, 4))

# adjust the position
pos = nx.spring_layout(G)

# draw the figure
nx.draw(G, pos, with_labels=False, node_color='red', node_size=25, font_weight='bold')

# save the figure
#plt.savefig('random graph.svg',format='svg')

plt.show()
```

### scale-free network ( $N = 100$, $k = 1$ )

```python
# initialize a scale-free network with 100 nodes, and the average degree is 1
ba = nx.barabasi_albert_graph(100, 1)

# adjust the position
pos = nx.spring_layout(ba)

# the size of figure
plt.figure(figsize=(4, 4))

# draw the figure
nx.draw(ba, pos, with_labels = False, node_color='red', node_size = 25)

# save the figure
#plt.savefig('scale-free graph.svg',format='svg')

plt.show()
```

### small-world network ( $N = 10$, $k = 5$ )

```python
# initialize a small-world network with 10 nodes, and the average degree is 5
small_world_graph = nx.watts_strogatz_graph(10, 5, 0.5)

# the size of figure
plt.figure(figsize=(4, 4))

# adjuest the position
pos = nx.circular_layout(small_world_graph)

# draw the figure
nx.draw(small_world_graph, pos, with_labels=False, node_color='red', node_size=25, font_weight='bold')

# save the figure
#plt.savefig('小世界网络（small-world graph）.svg',format='svg')

plt.show()
```
